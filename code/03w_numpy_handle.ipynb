### 3.2 내장함수 arange()와 linspace()

```python
import numpy as np

print(np.arange(3))
print(np.arange(3,7))
print(np.arange(3,7,2))
```

```python
import numpy as np

print(np.arange(2.0))
print(np.arange(2.0,6.5))
print(np.arange(2.0,6.5,0.8))
```

```python
a= np.arange(12).reshape(3,4)
a
```

```python
np.reshape(a,(2,6))
```

```python
np.reshape(a,(3,5))
```

```python
#함수 linspace()
np.linspace(2.0,3.0)
```

```python
np.linspace(2.0,3.0,retstep=True)
```

```python
print((3-2)/(50-1))
```

```python
np.linspace(2.0,3.0,endpoint=False,retstep=True)
```

```python
print((3-2)/(50))
```

```python
np.linspace(2.0,3.0,num=5,retstep=True)
```

```python
np.linspace(2.0,3.0,num=5,endpoint=False,retstep=True)
```
