### NumPy 배열 결합 함수 요약

| 함수명 | 설명 | 축 (Axis) | 특징 |
|:---:|:---|:---:|:---|
| **vstack** | 수직(Vertical) 방향 결합 | axis=0 | 배열을 아래로 쌓음 (행 추가) |
| **hstack** | 수평(Horizontal) 방향 결합 | axis=1 | 배열을 옆으로 붙임 (열 추가) |
| **row_stack** | 행(Row) 방향 결합 | axis=0 | vstack과 동일한 동작 (Alias) |
| **column_stack** | 열(Column) 방향 결합 | axis=1 | 1차원 배열을 열 벡터로 변환하여 결합 |

### 💡 핵심 차이점 (1차원 배열 처리)

* **hstack([a, b])**: 1D 배열을 단순 연결하여 **1차원** 결과 반환
* **column_stack([a, b])**: 1D 배열을 각각의 열로 취급하여 **2차원** 결과 반환

```python
x=np.arange(1,4)
y=np.arange(4,7)
x,y
```

```python
x=np.arange(1,7).reshape(2,3)
y=np.arange(7,13).reshape(2,3)
x,y
```
