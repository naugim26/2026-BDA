```python
from matplotlib import colors
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2025)
N=80
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
area=(20*np.random.rand(N))**2

plt.scatter(x,y,s=area,c=colors,alpha=0.7)
plt.colorbar()
plt.show()
```

```python
x=np.random.randint(-100,100,1000)
y=np.random.randint(-100,100,1000)

mask1=abs(x)>50
mask2=abs(y)>50
x=x[mask1+mask2]
y=y[mask1+mask2]

plt.scatter(x,y,c=x, cmap='jet',alpha=0.4)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.suptitle('전체 제목',fontsize=14, y=1.02)

plt.subplot(2,3,1)
plt.plot([1,4,9,16])
plt.title('제곱수')

plt.subplot(2,3,6)
plt.plot([1,2,3,4])
plt.title('선형')

plt.subplot(2,3,2)
plt.plot([6,5,2,1])
plt.title('감소')

plt.tight_layout() #중첩 방지
plt.show()
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots(2,3,figsize=(7,4))
fig.set_size_inches(6,3)

ax[0,1].plot
ax[1, 0].plot([1, 2, 3, 4])
ax[1, 2].plot([6, 5, 2, 1])

# 전체 제목 — Figure 레벨
# fig.suptitle('전체 제목', fontsize=14, fontweight='bold', y=1.02)
fig.suptitle('전체 제목', fontsize=14, fontweight='bold')

# 각 subplot 제목 — Axes 레벨
ax[0, 0].set_title('(0,0) 빈 플롯')
ax[0, 1].set_title('(0,1) 제곱수')
ax[0, 2].set_title('(0,2) 빈 플롯')
ax[1, 0].set_title('(1,0) 선형')
ax[1, 1].set_title('(1,1) 빈 플롯')
ax[1, 2].set_title('(1,2) 감소')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6, 4))

ax1 = fig.add_subplot(231)
ax1.plot([1, 4, 9, 16])
ax1.set_title('제곱수')

ax2 = fig.add_subplot(233)
ax2.plot([1, 2, 3, 4])
ax2.set_title('선형')

ax3 = fig.add_subplot(235)
ax3.plot([6, 5, 2, 1])
ax3.set_title('감소')

fig.suptitle('전체 제목', fontsize=14)

plt.tight_layout()
plt.show()
```
