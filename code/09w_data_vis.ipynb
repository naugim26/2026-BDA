| 선 스타일 이름 linestyle | 선 스타일 기호 | 예제 코드 |
|---|---|---|
| 실선(Solid line) | `'-'` | `plt.plot(x, y, linestyle='-')` |
| 점선(Dashed line) | `'--'` | `plt.plot(x, y, linestyle='--')` |
| 점점선(Dotted line) | `':'` | `plt.plot(x, y, linestyle=':')` |
| 대시 점선(Dash-dot line) | `'-.'` | `plt.plot(x, y, linestyle='-.')` |
| 없음(None) | `''` 또는 `'None'` | `plt.plot(x, y, linestyle='None')` |

```python
x = np.linspace(0, 10, 100)

# 다양한 선 스타일
styles = ['solid', 'dotted', 'dashed', 'dashdot', 'None']

for i, style in enumerate(styles):
    plt.plot(x, [i]*len(x), linestyle=style, linewidth=2, label=style)

plt.yticks(range(len(styles)), styles)

plt.xlim(0, 10)
plt.grid(True)

plt.show()
```
