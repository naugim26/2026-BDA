```python
import seaborn as sns
ax = sns.countplot(data = mtc, x='gear',
                   order = mtc.gear.value_counts().index)
for p in ax.patches:
  height = p.get_height()
  ax.text(x=(p.get_x() + p.get_width()/2.), y=(height+.2),
          s=int(height), ha='center',size=12,color='red')

ax.set_ylim(0,17)
plt.show()
```

```python
from pydataset import data
iris=data('iris')
iris.info()
```

```python
iris.Species.unique()
```

```python
iris.Species.value_counts()
```

```python
n, bins, patches = plt.hist(iris['Sepal.Length'],bins=20,
                            color='yellowgreen',edgecolor='white')
print(n)
print(bins)
print(patches)
```

```python
plt.hist(iris['Sepal.Length'],bins=20,
                            color='red',edgecolor='white',alpha=.5)
```

```python
plt.cm.viridis
```

```python
max(n)
```

```python
n, bins, patches = plt.hist(iris['Sepal.Length'],
                            bins=20,
                            color='red',
                            edgecolor='white')
for i in range(len(patches)):
  patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))
plt.show()
```
